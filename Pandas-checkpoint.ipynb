```python
import pandas as pd
```

```python
df = pd.read_excel('Retail 1.xlsx')
```

### Cleaning dirty data

```python
# handle missing values
df = df.dropna()
```

```python
df
```

```python
## remove outliers
df = df[(df['Quantity']>0)&(df['UnitPrice']>0)]
```

### Transforming data

```python
df['TotalPrice'] = df['UnitPrice'] * df['Quantity']
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```
